```python
import seaborn as sns
for column in df.columns:
    if column not in ['Cloud Cover','Season','Weather Type','Location']:
        plt.figure()
        sns.displot(df[column])
        plt.show()

        plt.figure()
        sns.displot(df[column], kind = "kde")
    else:
        plt.figure()
        sns.displot(df[column])
        plt.show()
```

```python

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder() 
categories =['Cloud Cover', 'Season', 'Location']
for col in categories:
    df[col] = label_encoder.fit_transform(df[col])
```

```python
print(df.head())
```

```python
X=df.drop(columns='Weather Type')
y=df['Weather Type']
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.25 , random_state=42)
print(X_train)
print(y_train)
print(X_test)
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
print(X_train)
print(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
classifier.score(X_train,y_train)
pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import classification_report

acc=accuracy_score(y_test, pred)
print("Accuracy:",acc)
print(classification_report(y_test , pred))
```

```python
cm = confusion_matrix(y_test, pred)
print(cm)
```

```python
classes = np.unique(y_test)
print(classes)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#ANN
y_train = pd.get_dummies(y_train)
y_test = pd.get_dummies(y_test)
```

```python
model = Sequential()


model.add(Dense(units=64, activation='relu', input_dim=X_train.shape[1]))

#hidden:
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=16, activation='relu'))
```

```python
model.add(Dense(units=4, activation='softmax'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=53, batch_size=32, validation_data=(X_test, y_test))
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print("Loss: ",loss)
print("Accuracy:",accuracy)
```

```python
y_pred = model.predict(X_test)
y_pred_cla = np.argmax(y_pred, axis=1)
y_test_cla = np.argmax(y_test.values, axis=1)

cm = confusion_matrix(y_test_cla, y_pred_cla)
print(cm)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt="d", cmap='Blues', xticklabels=np.unique(df['Weather Type']), yticklabels=np.unique(df['Weather Type']))
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
acc = accuracy_score(y_test_cla, y_pred_cla)
print(f"Accuracy: {acc}")
```

```python
print(classification_report(y_test_cla, y_pred_cla))
```
